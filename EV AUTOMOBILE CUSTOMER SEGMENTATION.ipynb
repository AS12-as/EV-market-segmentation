```python
# Prepare Data for Clustering and segmentation
EV_var['Profession'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
dfle= EV_var
dfle.Gender= le.fit_transform(dfle.Gender)
dfle
```

```python
dfle1= dfle
dfle1.Married= le.fit_transform(dfle1.Married)
dfle1
```

```python
dfle2= dfle1
dfle2.Profession= le.fit_transform(dfle2.Profession)
dfle2
```

```python
dfle3= dfle2
dfle3.SpendingScore= le.fit_transform(dfle3.SpendingScore)
dfle3
```

```python
dfle4= dfle3
dfle4.Graduated= le.fit_transform(dfle3.Graduated)
dfle4
```

```python
dfle4
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(dfle4)
```

```python
kmeans.labels_
```

```python
kmeans.predict
```
